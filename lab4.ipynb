```python
from qiskit import QuantumCircuit 
qc = QuantumCircuit(2)
qc.h(0) 
qc.cx(0, 1) 
qc.measure_all() 
print(qc.draw())
```

```python
from qiskit.primitives import StatevectorSampler
sampler = StatevectorSampler()
job = sampler.run([qc],shots=1000)
result = job.result()
print(f"> Counts: {result[0].data.meas.get_counts()}")
#bell state 00 , 11
```
