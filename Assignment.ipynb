```python
#Inner Product
import numpy as np
def inner_product(vec1, vec2):
    res = 0
    for i in range(vec1.size):
        res+= np.conjugate(vec1[i])*vec2[i]
    return res
v1 = np.array([1+2j, 3-1j, 1j])
v2 = np.array([2, 1+4j, 3+3j])
print(inner_product(v1, v2))
```

```python
#Norm
def vector_norm(vec):
    sum = 0
    for i in vec:
        sum+= np.conjugate(i) * i
    return np.real(np.sqrt(sum))
v = np.array([3+4j, 1-2j, 1j])
print(vector_norm(v))
```

```python
#Distance
def vectors_dis(vec1, vec2):
    diff = vec1 - vec2
    sum = 0
    for i in diff:
        sum += np.conjugate(i) * i
    return np.real(np.sqrt(sum))
print(vectors_dis(v1, v2))
```

```python
#hermitian
def is_hermitian(matrix):
    return np.allclose(matrix, np.conjugate(matrix.T))
A = np.array([[1, 2+1j],[2j, 3]])
print("is Hermitian?", is_hermitian(A))
```
