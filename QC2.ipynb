```python
import numpy as np


def add_vectors(v1, v2):
    return v1 + v2

def inverse_vector(v):
    return -v

def scalar_multiply_vector(s, v):
    return s * v

def add_matrices(m1, m2):
    return m1 + m2

def inverse_matrix(m):
    return -m

def scalar_multiply_matrix(s, m):
    return s * m

def multiply_matrices(m1, m2):
    return np.dot(m1, m2)
    # return m1 @ m2

def vector_matrix_action(v, m):
    return np.dot(v, m)



v1 = np.array([1+2j, 3+4j])
v2 = np.array([2+1j, 1+3j])
m1 = np.array([[1+2j, 2+3j], [3+1j, 4+2j]])
m2 = np.array([[2+1j, 1+4j], [0+2j, 3+3j]])
s = 2+1j

print("Vector Addition:", add_vectors(v1, v2))
print("Vector Inverse:", inverse_vector(v1))
print("Scalar * Vector:", scalar_multiply_vector(s, v1))
print("\nMatrix Addition:\n", add_matrices(m1, m2))
print("\nMatrix Inverse:\n", inverse_matrix(m1))
print("\nScalar * Matrix:\n", scalar_multiply_matrix(s, m1))
print("\nMatrix Multiplication:\n", multiply_matrices(m1, m2))
print("\nVector-Matrix Action:\n", vector_matrix_action(v1, m1))
```
