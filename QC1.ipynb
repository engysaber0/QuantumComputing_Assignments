```python

import cmath

a1 = int(input())
a2 = int(input())
a3 = int(input())
a4 = int(input())

c1 = complex(a1,a2)
c2 = complex(a3,a4)

print(c1*c2)
print(c1+c2)
print(c1/c2)
print(c1-c2)
print(c1.conjugate())

r , theta = cmath.polar(c1)
r1 , theta1 = cmath.polar(c2)

cartesian = cmath.rect(r,theta)
print(cartesian)
cartesian1 = cmath.rect(r1,theta1)
print(cartesian1)
```
